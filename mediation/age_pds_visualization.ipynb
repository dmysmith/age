```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
from matplotlib import rc
import pylab
from pylab import * 
```

```python
import seaborn as sns
```

```python
# load data
ndafile = '/space/syn50/1/data/ABCD/d9smith/age/nda5.0_bfs.txt'
data = pd.read_csv(ndafile, delimiter='\t')
```

```python
# data["age"] = data["interview_age"]/12
data["age"] = pd.to_numeric(data["interview_age"]/12)
```

```python
plotdf = data.dropna()
```

# Average of Parent and Youth Report PDS

```python
# scatterplot of age and PDS
color_dict = { 'F':'pink', 'M':'blue', 'I':'green'}
fig = plt.figure()
fig.subplots_adjust(bottom=0.2)
ax = fig.add_subplot(111)
plt.scatter(plotdf.age,plotdf.pds_y_p_average,alpha=0.2, c=[color_dict[i] for i in plotdf.sex])
```

```python
sns.swarmplot(data=plotdf
              , x="age"
              , y="pds_y_p_average"
              , hue="sex"
              , dodge=True
              , size=2
              , orient = "h"
              , alpha = 0.5
              , native_scale=True)
```

```python
# take random sample of 1000
import random
idx=random.sample(list(range(1,data.shape[0])),1000)
data_random = data.iloc[idx]
```

```python
# scatterplot of age and PDS in random 1000
fig = plt.figure()
fig.subplots_adjust(bottom=0.2)
ax = fig.add_subplot(111)
plt.scatter(data_random.dropna().age,data_random.dropna().pds_y_p_average,alpha=0.2, c=[color_dict[i] for i in data_random.dropna().sex])
```

```python
# use a handmade palette
# pinkblue = ["pink", "lightblue"]
pinkblue = ["cornflowerblue", "hotpink", "green"]
sns.set_palette(pinkblue)
```

```python
sns.swarmplot(data=data_random
              , x="age"
              , y="pds_y_p_average"
              , hue="sex"
              , dodge=True
              , size=2
              , orient = "h"
              , native_scale=True)
```

```python
sns.stripplot(data=data_random
              , x="age"
              , y="pds_y_p_average"
              , hue="sex"
              , orient = "h"
              , jitter=True
              # , alpha=0.9
              , size=4
              , native_scale=True)
```
