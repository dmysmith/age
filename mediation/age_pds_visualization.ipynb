```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from numpy import *
from matplotlib import rc
import pylab
from pylab import * 
%matplotlib inline
```

```python
import seaborn as sns
```

```python
# load data
ndafile = '/space/syn50/1/data/ABCD/d9smith/age/nda5.0_bfs.txt'
data = pd.read_csv(ndafile, delimiter='\t').dropna()
```

```python
# data["age"] = data["interview_age"]/12
data["age"] = pd.to_numeric(data["interview_age"]/12)
```

```python
plotdf = data.dropna()
```

# Average of Parent and Youth Report PDS

```python
# scatterplot of age and PDS
color_dict = { 'F':'pink', 'M':'lightblue', 'I':'green'}
fig = plt.figure()
fig.subplots_adjust(bottom=0.2)
ax = fig.add_subplot(111)
plt.scatter(data.age,data.pds_y_p_average,alpha=0.2, c=[color_dict[i] for i in plotdf.sex])
```

```python
# take random sample of 1000
import random
idx=random.sample(list(range(1,data.shape[0])),1000)
data_random = data.iloc[idx]
```

```python
# scatterplot of age and PDS in random 1000
fig = plt.figure()
fig.subplots_adjust(bottom=0.2)
ax = fig.add_subplot(111)
plt.scatter(data_random.age,data_random.pds_y_p_average,alpha=0.2, c=[color_dict[i] for i in data_random.sex])
```

# Comparison of youth vs. parent report PDS

```python
# data_random['pds_p_ss_category_all'] = data_random.pds_p_ss_category_all.astype(str)
# data_random['pds_y_ss_category_all'] = data_random.pds_y_ss_category_all.astype(str)

# data_random['pds_p_ss_category_all'] = data_random['pds_p_ss_category_all'].astype(int)
# data_random['pds_y_ss_category_all'] = data_random['pds_y_ss_category_all'].astype(int)

pds_p_ss_category = data_random['pds_p_ss_category_all'].astype(int)
pds_y_ss_category = data_random['pds_y_ss_category_all'].astype(int)

enc_data = pd.concat([pds_p_ss_category, pds_y_ss_category], axis=1)
enc_data.head()
```

```python
# This is to encode the data into numbers that can be used in our scatterplot
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
enc_df = pd.DataFrame(ord_enc.fit_transform(enc_data), columns=list(enc_data.columns))
categories = pd.DataFrame(np.array(ord_enc.categories_).transpose(), columns=list(enc_data.columns))

# Generate the random noise
xnoise, ynoise = np.random.random(len(enc_data))/2, np.random.random(len(enc_data))/2 # The noise is in the range 0 to 0.5

# Plot the scatterplot
plt.scatter(enc_df["pds_p_ss_category_all"]+xnoise, enc_df["pds_y_ss_category_all"]+ynoise, alpha=1,s=2,c=[color_dict[i] for i in data_random.sex])
plt.xlabel("pds_p_ss_category_all")
plt.ylabel("pds_y_ss_category_all")
# You can also set xticks and yticks to be your category names:
plt.xticks([0.25, 1.25, 2.25, 3.25, 4.25], categories["pds_p_ss_category_all"]) # The reason the xticks start at 0.25
# and go up in increments of 1 is because the center of the noise will be around 0.25 and ordinal
# encoded labels go up in increments of 1.
plt.yticks([0.25, 1.25, 2.25, 3.25, 4.25], categories["pds_y_ss_category_all"]) # This has the same reason explained for xticks

# Extra unnecessary styling...
plt.grid()
sns.despine(left=True, bottom=True)
```
