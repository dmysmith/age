```r
for (p in c("pracma", "splines", "Matrix")){
        if(!eval(parse(text=paste("require(",p,")")))) {
                install.packages(p)
                lapply(p,library,character.only=TRUE)
        }
}
```

```r
xvec = linspace(100,200,101)
```

```r
basis = ns(xvec, knots = c(120,140,160,180), intercept = FALSE)
dim(basis)
rankMatrix(basis)

par(mfrow=c(2,3))
plot(xvec, basis[,1]) 
plot(xvec, basis[,2]) 
plot(xvec, basis[,3]) 
plot(xvec, basis[,4])
plot(xvec, basis[,5]) 
```

```r
basis = data.frame(ns(xvec, knots = c(120,140,160,180), intercept = FALSE))
basis_demeaned = sapply(basis,function(x)x-mean(x))
dim(basis_demeaned)
rankMatrix(basis_demeaned)
                        
par(mfrow=c(2,3))
plot(xvec, basis_demeaned[,1]) 
plot(xvec, basis_demeaned[,2]) 
plot(xvec, basis_demeaned[,3]) 
plot(xvec, basis_demeaned[,4])
plot(xvec, basis_demeaned[,5]) 
```

```r
basis = ns(xvec, knots = c(120,140,160,180), intercept = TRUE)
dim(basis)
rankMatrix(basis)

par(mfrow=c(2,3))
plot(xvec, basis[,1]) 
plot(xvec, basis[,2]) 
plot(xvec, basis[,3]) 
plot(xvec, basis[,4])
plot(xvec, basis[,5]) 
plot(xvec, basis[,6])
```
